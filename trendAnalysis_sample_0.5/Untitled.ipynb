```python
%matplotlib qt
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("part-00000.csv",header=None)
df=df.rename(columns={0: 'Date'})
df['Date'] =pd.to_datetime(df.Date)
df=df.sort_values(by=['Date'])
df.set_index('Date', inplace=True)
df.head()

df.plot(subplots=True)

plt.tight_layout()
plt.show()
```

```python
with open("topics.txt",encoding="utf8") as file:
    lines = file.readlines()
    topics = [line.rstrip().split(", ") for line in lines]
    for i in range(0,len(topics)):
        exclude=i
        not_in_others=topics[i]
        for t in range(0,len(topics)):
            if t==exclude:
                continue
            not_in_others=list(set(not_in_others) - set(topics[t]))
        print(str(i+1)+' : '+str(not_in_others))
```
